```python
import pandas as pd

```

```python
df = pd.read_csv("Data/athlete_events.csv")
df.head()
```

```python
import seaborn as sns

# can also use groupby with size
male_female_count = (
    df[["Year", "Sex"]].value_counts().reset_index().sort_values(by="Year")
)

sns.lineplot(male_female_count, x="Year", y="count", hue="Sex")
```
